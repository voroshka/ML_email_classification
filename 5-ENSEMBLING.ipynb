```python
import pandas as pd
import numpy as np
```

```python
rf = pd.read_csv("../data/submissions/rf_probas.csv")
xgb = pd.read_csv("../data/submissions/xgb_probas.csv")
lgb = pd.read_csv("../data/submissions/lgbm_probas.csv")
```

```python
for df in [rf, xgb, lgb]:
    df.set_index("Id", inplace=True)
```

```python
ensemble = (rf + xgb + lgb) / 3
```

```python
ensemble
```

```python
submission = pd.DataFrame()
submission["Id"] = range(len(rf))
```

```python
submission["label"] = ensemble.values.argmax(axis=1)
```

```python
submission
```

```python
submission.label.value_counts()
```

```python
submission.to_csv("../data/submissions/ensemble.csv", index=False)
```
